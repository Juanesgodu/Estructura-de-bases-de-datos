# Ejercicio árbol Hash

```python
import hashlib


text0= "hola"
text1= "Hola"
text2= "chao"
text3= "Chao"


def hash(text):
    hash_object = hashlib.sha256(text.encode())
    return hash_object.hexdigest()


def tree(text0,text1, text2, text3):

    hash0= hash(text0)
    hash1= hash(text1)
    hash2= hash(text2)
    hash3= hash(text3)

    hash01= hash(hash0 + hash1)
    hash23= hash(hash2 + hash3)

    root_hash= hash(hash01 + hash23)
    return root_hash


print(tree(text0,text1, text2, text3))


#re= hash(text0)
#print(re)
```

# Ejercicio árbol de Merkle

```python
# proporcionado por Chatgpt

import hashlib

class MerkleTree:

    def __init__(self, transacciones):
        self.transacciones = transacciones
        self.root = self.construir_arbol()

    def hash(self, data):
        return hashlib.sha256(data.encode()).hexdigest()

    def construir_arbol(self):
        nivel = [self.hash(tx) for tx in self.transacciones]

        while len(nivel) > 1:

            if len(nivel) % 2 != 0:
                nivel.append(nivel[-1])
            print(nivel)
            nuevo_nivel = []

            for i in range(0, len(nivel), 2):
                combinado = nivel[i] + nivel[i+1]
                nuevo_nivel.append(self.hash(combinado))

            nivel = nuevo_nivel

        return nivel[0]


# Uso
transacciones = ["hola", "Hola", "chao", "Chao", "xd"]
merkle = MerkleTree(transacciones)

print("Merkle Root:", merkle.root)
```
